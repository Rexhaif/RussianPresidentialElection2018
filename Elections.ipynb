```python
def find_uik(region, tik, uik):
    region_uik = table.loc[table['region_name'] == region]
    region_tik = region_uik.loc[region_uik['tik_name'] == tik]
    result = region_tik.loc[region_tik['uik_name'] == uik]
    if (len(result) == 0):
        return "Error, try again"
    result = result.as_matrix()
    result = result[0]
    result = pd.Series(result, index=list(table))
    return result

def find_uik_w_tik(region, uik):
    region_uik = table.loc[table['region_name'] == region]
    result = region_uik.loc[region_uik['uik_name'] == uik]
    if (len(result) == 0):
        return 0
        #return "Does not exist, try again"
    if (len(result) > 1):
        return -1
        #print("More than 1 candidate, select appropriate TIK and use find_uik")
        #result = result['tik_name']
        #return result
    result = result.as_matrix()
    result = result[0]
    result = pd.Series(result, index=list(table))
    return result

def find_o_uik(uik):
    result = table.loc[table['uik_name'] == uik]
    if (len(result) == 0):
        return 0
        #return "Does not exist, try again"
    if (len(result) > 1):
        return -1
        #print("More than 1 candidate, select appropriate region and use find_uik_w_tik")
        #result = result['region_name'].drop_duplicates()
        #return result
    result = result.as_matrix()
    result = result[0]
    result = pd.Series(result, index=list(table))
    return result
```

```python
def select_uik():
    pd = table['region_name'].drop_duplicates()
    pd = pd.as_matrix()
    print('All existing regions:', pd)    
    reg = input('Enter your region: ')
    region_uik = table.loc[table['region_name'] == reg]
    if (len(region_uik) == 0):
        return "Error, try again"
    pd = region_uik['tik_name'].drop_duplicates()
    pd = pd.as_matrix()
    print('All existing TIKs:', pd)
    tik = input('Enter your TIK: ')
    region_tik = region_uik.loc[region_uik['tik_name'] == tik]
    if (len(region_tik) == 0):
        return "Error, try again"
    pd = region_tik['uik_name'].drop_duplicates()
    pd = pd.as_matrix()
    print('All existing UIKs:', pd)
    uik = input('Enter your UIK: ')
    return find_uik(reg, tik, uik)
```
